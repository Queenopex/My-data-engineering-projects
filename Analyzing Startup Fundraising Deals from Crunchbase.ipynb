The memory usage of each column

```python
chrunchbase = pd.read_csv('crunchbase-investments.csv', 
                          chunksize = 5000, encoding='ISO-8859-1')
mem_use = []
for chunk in chrunchbase:
    mem = chunk.memory_usage(deep = True).sum()/ 1024**2
    mem_use.append(mem)
    
print(mem_use)
```

```python
chrunchbase = pd.read_csv('crunchbase-investments.csv', 
                          chunksize = 5000, encoding='ISO-8859-1')
value = []
for chunk in chrunchbase:
    value.append(chunk.isnull().sum())
    
combined_mv = pd.concat(value)
unique_combined_mv = combined_mv.groupby(combined_mv.index).sum()
unique_combined_mv.sort_values()
```

```python
col_types = {}
chunk_iter = pd.read_csv('crunchbase-investments.csv',
chunksize = 5000, encoding='ISO-8859-1', usecols = keep_col)
for chunk in chunk_iter:
    for col in chunk.columns:
        if col not in col_types:
            col_types[col] = [str(chunk.dtypes[col])]
        else:
            col_types[col].append(str(chunk.dtypes[col]))
            
            

for col, types in col_types.items():
    most_common_type = max(set(types), key=types.count)
    col_types[col] = most_common_type

# Print the inferred data types for each column
print(col_types)
        
```

Loading Chunks into SQLite

```python
import sqlite3
conn = sqlite3.connect('crunchbase.db')

chunk_iter = pd.read_csv('crunchbase-investments.csv', chunksize = 50000,
                        encoding = 'ISO-8859-1')
for chunk in chunk_iter:
    chunk.to_sql("investments", conn, if_exists = 'append', index = False)
```
